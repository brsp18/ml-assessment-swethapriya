```python

df.isna().sum()

#total_missing_values= df.isna().sum().sum()
#print("Total Missing values", total_missing_values,"/n")

df['resting_bp']= df['resting_bp'].fillna(df['resting_bp'].median())
df['cholesterol']= df['cholesterol'].fillna(df['cholesterol'].median())
print("New Dataframe after filling null values:")
print( df.isnull().sum())
print(f"Total rows after: {len(df)}")
```

As noted in the begening, Only two columns had missing values:
- `resting_bp`: 24 missing values (3%)
- `cholesterol`: 32 missing values (4%)


I opted Median imputation was over row deletion for the following reasons:

1.Only 3-4% of values are missing in the columns. Dropping the row will effect 56 patients wich is almost 7% of the datset which I feel is not a recommended option.

2. I opted Median over mean because these two columns have outliers which are expected to  deviate the data values. In such scenarios, Median is strongly recommended.

3. This data set is small with just 800 rows. So it is preferred to ratain the records rather than dropping them.

2.resting_ecg: 3 unique values — ['left_ventricular_hypertrophy', 'st_t_wave_abnormality', 'normal']

3.st_slope: 3 unique values — ['up', 'flat', 'down']

One hot encoding resulted in dataset expansion from 12 columns to 19 columns with all features in numerical standards ready for model training
